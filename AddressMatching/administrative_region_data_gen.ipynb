## Build region tree

```python
import re

def remove_prefix_from_region_name(prefixed_name: str, prefix_list: [str]) -> (str, str):
    
    for prefix in prefix_list:
        p = re.compile(prefix)
        m = p.match(prefixed_name)
        if m is None:
            continue
        return (prefix, prefixed_name[m.end():].strip())
        
    return ("", prefixed_name)
```

```python
province_prefix = ["Thành phố", "Tỉnh"]
district_prefix = ["Quận", "Huyện", "Thị xã", "Thành phố"]
ward_prefix = data[ward_level_prefix_colname].unique()
```

```python
def add_region_to_tree(tree, prov_prefix, prov_name, dist_prefix, dist_name, ward_prefix, ward_name) -> None:
    
    if prov_name not in tree:
        tree[prov_name] = {'prefix': prov_prefix, 'name': prov_name, 'districts': {}}
        
    dists = tree[prov_name]['districts']
    if dist_name not in dists:
        dists[dist_name] = {'prefix': dist_prefix, 'name': dist_name, 'wards': {}}
        
    wards = dists[dist_name]['wards']
    if ward_name not in wards:
        wards[ward_name] = {'prefix': ward_prefix, 'name': ward_name}
```

```python
region_tree = {}

for i in data.index:
    row = data.loc[i]
    prov = row[province_colname]
    dist = row[district_colname]
    ward = row[ward_colname]
    ward_prefix = row[ward_level_prefix_colname]
    
    prov_prefix, prov_name = remove_prefix_from_region_name(prov, province_prefix)
    dist_prefix, dist_name = remove_prefix_from_region_name(dist, district_prefix)
    ward = ward.removeprefix(ward_prefix).strip()
    
    add_region_to_tree(region_tree, prov_prefix, prov_name, dist_prefix, dist_name, ward_prefix, ward)
```
